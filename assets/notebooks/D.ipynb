```python
#Q1
#Simple Calculator

#collecting the operation to be perfromed from the user
operation_type = input('Please enter the operation you want to perfrom(between +,-,* and /):')

#collecting the operands from the user
operand1 = float(input('Enter the first operand:'))
operand2 = float(input('Enter the second operand:'))

#performing operation
if operation_type == '+':
    print(f'{operand1} + {operand2} = {operand1 + operand2}')
elif operation_type == '-':
    print(f'{operand1} - {operand2} = {operand1 - operand2}')
elif operation_type == '*':
    print(f'{operand1} X {operand2} = {operand1 * operand2}')
elif operation_type == '/':
    print(f'{operand1} / {operand2} = {operand1 / operand2}')
else:
    print('invalid operator')
```

```python
#Q2
#BMI Calculator

#Collecting user inputs to calculate BMI
mass = float(input('Please input your weight(in kilograms):'))
height = float(input('Please input your height(in metres):'))

#Using furmula to calculate BMI
BMI = mass/(height ** 2)

#using if elif and else to determing wether the inputs belong to an underweight, overweight or normal weight person
if BMI <= 18.5:
    print('you are underweight')
elif BMI <= 25:
    print('your weight is normal')
else:
    print('you are overweight')
```

```python
#Q3

#a)Extracting unique dictionary values from a given dictionary
def get_unique_values(dic):    
    #using the .values function to get a list of the values in the dictionary, then converting it into a set to remove duplicates
    unique_values = set(dic.values())
    
    return unique_values

#b)Happy numbers
def is_happy(num):
    
    #while loops that exits if the number is 1,2 or three from which only those that reach 1 are happy numbers 2 and 3 run infinitely into themselves
    while num >= 10 :
        
        #storing the number as a string to access each digit
        n = str(num)
        
        #changing num to 0 so it can be updated as the sum of all the digits of num
        num = 0
        
        #using a for loop to access every element in num to square it to make the new num
        for i in n:
            num += int(i) ** 2
        
    #the number is happy if num = 1 or 6therefore we return the output of num == 1 or num == 7 as they are the only happy numbers below 10
    return num == 1 or num == 7
        
#Tests
#a) unique values test
dic = {1:1 , 2:1, 3:1, 4:2, 5:1, 6:3, 7:1}
print(get_unique_values(dic))

#b) happy numbers test
print(is_happy(23))
```
