```python
# QUESTION 1


# Inputting the type of operation wanted as well as the operands required
operation = input("Enter Operation of Your Choice (Addition(+) / Subtraction(-) / Multiplication(x) / Division(/)):")
numbers = input("Enter 2 Numbers/Operands with a Space:")

# Splitting and assigning the numbers to their respective variables
n1,n2 = numbers.split()
n1 = float(n1)
n2 = float(n2)

operation = operation.lower() # Makes the the user input for operation lowercase


# To check the value of the operation input which will accordingly print the chosen operation on the inputted numbers

if (operation == 'addition') or (operation == '+'):           # Checks the operation and prints the sum of the numbers if it's 'addition'
    print(f'Sum = {n1 + n2}')
elif (operation == 'subtraction') or (operation == '-'):      # Checks the operation and prints the difference of the numbers if it's 'subtraction'
    print(f'Difference = {n1 - n2}')
elif (operation == 'multiplication') or (operation == 'x'):   # Checks the operation and prints the multiplication of the numbers if it's 'multiplication'
    print(f'Multiplication = {n1 * n2}')
elif (operation == 'division') or (operation == '/'):         # Checks the operation and prints the division of the numbers if it's 'division'
    print(f'Division = {n1 / n2}')
else:
    print('Invalid Input')                        # Prints Invalid Input if the user input for operation is not valid
```

```python
# QUESTION 2

# To input the mass in kilograms and height in meters
mass = float(input('Enter your Weight in Kilograms:'))
height = float(input('Enter your Height in Meters:'))

# Using the website's formula to calculate Body Mass Index(BMI)
bmi = mass / (height**2) 

# Checking the relevant conditions to categorize weight as underweight, normal or overweight

if bmi <= 18.5:                      # Checks the BMI and prints 'Underweight' if it's less than/equal to 18.5
    print('You are Underweight.')
elif (18.5 < bmi) and (bmi <= 25):   # Checks the BMI and prints 'Normal' if it's over 18.5 and less than/equal to 25
    print('Your Weight is Normal.')
elif bmi > 25:                       # Checks the BMI and prints 'Overweight' if it's over than 25
    print('You are Overweight.')
```

```python
# QUESTION 3

# a) Extract Unique Values from a Dictionary

# Creating a dictionary with different keywords and values
dic = {'a':'shirt','b':'pants','c':'shirt','d':'socks','e':'pants','f':'shoes','g':'socks'} # Created Dictionary
unique = set() # Initializing an empty set

# For loop to add all values to the empty set
for uniq_val in dic.values():
    unique.add(uniq_val)

# Prints the unique values in the dictionary
print('Unique Values in Dictionary:',unique)



# b) Find if a Number is a Happy Number or Not

# To get user input of a number
num = int(input('Enter a Number to Check if Happy or Not:'))

# Function to return the sum of squares of the digits

def isHappy(n):
    powersum = 0
    while n > 0:
        d = n % 10           # To extract the last digit after each loop of the digit
        powersum += d**2     # To add and store the sum of the square of each digit of the number
        n //= 10             # To remove the last digit from the number after square of that digit is calculated
        
        if (n == 0) and (powersum < 10):       # To break the loop if the sum is a single digit number
            break
        elif (n == 0) and (powersum >= 10):    # To continue the loop and calculate the squares of the result if the sum is a double digit number
            n = powersum          # To assign the result as the new number
            powersum = 0          # To reset the variable holding sum of squares of the digits to 0 to recalculate new number
    return powersum

# To print result of the program

if isHappy(num) == 1:                        # Prints "is a Happy Number" if result is 1
    print(f'{num} is a Happy Number')
else:                                        # Prints "is not a Happy Number" if result is not 1
    print(f'{num} is NOT a Happy Number')
```
