```python
# Object: Simple Calculator
# Obtaining Input from the user.
operator = input("Please enter the operation to perform (+, -,*,/):")
operands = input("Please enter two operands separated by space:")

# Separating the input to obtain two operands.
operand1, operand2 = operands.split()

# Enabling computation of inputted operands.
operand1 = float(operand1)
operand2 = float(operand2)

# Calculating and printing the result based on the input.
for x in operator:
  if x == "+" :
    print(f"The sum is: {operand1 + operand2}")
  elif x == "-" :
    print(f"The difference is: {operand1 - operand2}")
  elif x == "*" :
    print(f"The product is: {operand1 * operand2}")
  else:
    print(f"The quotient is: {operand1 / operand2}")
```

```python
# Object: A BMI calculator which returns categorisation.
# Obtaining input in float format.
height = float(input("Please enter your height in meters:"))
weight = float(input("Please enter your weight in kilograms:"))

# Calculating BMI.
bmi = (weight
       /(height*height))

# Printing the result based on the categorisation.
if bmi <= 18.5:
  print("You are underweight")
elif bmi <= 25.0:
  print("Your weight is normal")
else:
  print("You are overweight")
```

```python
# Defining a function to give sum of square of each digit of a number
def sum_square(numbers):
  """
  Gives the sum of square of each digit of any number.
  """
  digits = [int(d) for d in str(numbers)]
  sqr = [x**2 for x in digits]
  sum_ = sum(sqr)
  return sum_

# Defining a function to evaluate if a number is a happy number or not
def happy_num(n):
  """
  Evaluates whether the given number is a happy number or not
  """
  number = int(n)
# All unhappy numbers eventually fall in the loop of 4, 16, 37, ...20.
  while number != 1 and number != 4:
    number = sum_square(number)
# Printing the categorised result.
  if number == 1:
    print(n, "is a happy number")
  else:
    print(n, "is not a happy number")

#Testing the function with two function calls.
happy_num(85)
happy_num(44)
```
